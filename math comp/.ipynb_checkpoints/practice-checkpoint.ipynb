## Generate a random normal distribution of size of size 2*3

```python
from numpy import random
```

```python
x = random.normal(size=(2,3))
print(x)
```

## Generate a normal distribution of 2*3 with mean at 1 and standard deviation of 2

```python
from numpy import random
```

```python
x=random.normal(loc=1,scale=2,size=(2,3))#scale is the same as standard deviation and loc is mean
print(x)
```
