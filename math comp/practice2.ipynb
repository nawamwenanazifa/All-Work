```python
from numpy import random
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

# Qn1.

# Given ten trials for coin tose generate the 10 data points

```python
x=random.binomial(n=10, p=0.5, size=10)
print(x)
```

# Visualization of binomial distribution

```python
binomial_data=random.binomial(n=10, p=0.5, size=1000)
sns.histplot(binomial_data, bins=10, kde=False, color="tomato")
plt.xlabel("Number of success")
plt.ylabel("Frequency")
plt.title("Visualization of Binomial Distribution")
plt.show()
```

# Difference between binomial distribution and normal distribution

```python
normal_data=random.normal(loc=50, scale=5, size=1000)
binomial_data=random.binomial(n=100, p=0.5, size=1000)
sns.kdeplot(normal_data, label="normal", color="slateblue")
sns.kdeplot(binomial_data, label="binomial", color="red")
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Difference between binomial distribution and normal distribution")
plt.legend()
plt.show()
```

# Poisson Distribution

```python
x=random.poisson(lam=2,size=10)#lam is number of occurance
print(x)
```

# Visualization of poisson distribution


```python
poisson_data=random.poisson( size=10)
sns.kdeplot(poisson_data, color="green")
plt.xlabel("Number of success")
plt.ylabel("Frequency")
plt.title("Visualization of Poisson Distribution")
plt.show()
```

# Uniform Distribution


# Generate 2 by 3 uniform distribution

```python
x=random.uniform(size=(2,3))
print(x)
```

# Visualization of uniform distribution

```python
uniform_data=random.uniform( size=1000)
sns.kdeplot(uniform_data, color="blue")
plt.xlabel("Number of success")
plt.ylabel("Frequency")
plt.title("Visualization of Uniform Distribution")
plt.show()
```
