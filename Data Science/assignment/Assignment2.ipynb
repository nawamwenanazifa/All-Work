```python
sns.set(style="whitegrid")
plt.figure(figure=(10,15))
sns.scatterplot(data=adult_df, x ='age', y= 'capitalGain', hue ="race",style="race", s=100)
plt.title('Scatterplot of Age vs Capital Gain')
plt.xlabel('Age')
plt.ylabel('Capital Gain')
plt.legend(title="Race")
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figure=(10,15))
sns.histplot(adult_df["age"], bins=20, color="red")
plt.title("Histogram of Age")
plt.xlabel('Age')
plt.ylabel("freguency")
plt.show()
```

```python
sns.countplot(x="race", data=adult_df)
plt.title("Race Distribution")
plt.xlabel("Race")
plt.ylabel("Frequency")
plt.show()
```
